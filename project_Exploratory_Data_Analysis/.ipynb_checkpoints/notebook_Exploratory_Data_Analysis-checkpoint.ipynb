### Определите и изучите пропущенные значения

Стоит сказать, что грамотно будет проверить каждый столбец на предмет очевидно неверных значений (типа отрицательных в столбце количества каких-либо объектов) и все объявления на предмет дубликатов. Однако, в плане проекта эти пункты опущены, поэтому считаем данные корректными, будем анализировать лишь столбцы с пропусками. Начнём с *ceiling_height* — высота потолков в метрах.

```python
print(apts['ceiling_height'].describe())
```

Я считаю, что пропуски в этом столбце возникли из-за простого незнания собственниками всех размеров своей квартиры, в том числе и высоты потолка.
Просматривая названия столбцов в данных, можно сказать, что нет очевидного варианта, как заменить пропуски. Заменить средним или медианным значением будет некорректно и грубо. Оставим пропуски незаполненными. Следом идет столбец *floors_total* — всего этажей в доме.

```python
print(apts['floors_total'].describe())
```

Довольно странно для тех, кто писал объявление, не включить в него этажность здания. Возможно, пропустили строку с количеством этажей в здании при заполнении данных. Или все же пропуски — это частные одноэтажные или двухэтажные дома, и жильцы не видели особого смысла в уточнениях. Не совсем ясно, почему одна замена пропуска будет лучше другой. Можно было бы привязать зависимость и от других переменных (например, расстояние от центра города), но на данном этапе анализа не будем этого делать. Далее *living_area* — жилая площадь в квадратных метрах.

```python
print(apts['living_area'].describe())
```

Пропусков немного, поэтому причиной тоже может служить либо невнимательность при написании, либо же незнание площади именно жилой части квартиры. Самым логичным решением мог быть подсчёт доли жилой площади в общем площади по всем объявлениям, а затем замена пропусков как раз общей площадью, умноженной на высчитанную долю. Но пока перейдем к *is_apartment* — столбец, показывающий, является ли квартира апартаментами, то есть является ли она нежилым помещением с точки зрения закона.

```python
print(apts['is_apartment'].value_counts())
```

Причина пропусков очевидна — продавец понимает, что не написав ничего, все поймут, что его жилье апартаментами не является. Здесь мы можем все пропуски заменить на False, то есть на то, что квартира апартаментами не является, а является частью жилого фонда.

```python
apts['is_apartment'] = apts['is_apartment'].fillna(False)
print(apts['is_apartment'].value_counts())
```

Итак, теперь всё заменили. *kitchen_area* — площадь кухни в квадратных метрах.

```python
print(apts['kitchen_area'].describe())
```

Как и в случае с жилой площадью, можно, опираясь на связь с другими столбцами, заполнить пропуски. Но этот вариант не самый очевидный и, возможно, не очень точный. Оставим пропуски незаполненными. Причиной пропусков можно считать то, что некоторые продавцы не знают площади кухни. Это нормально с учётом того, что некоторые не написали размер жилой площади квартиры. Теперь все пропуски в столбце заполнены. Столбец *locality_name* несёт в себе информацию о названии населённого пункта.

### Приведите данные к нужным типам

Среди уже рассмотренных столбцов нужно изменить тип данных у следующих:

* *floors_total* — всего этажей в доме (с float64 на int64, из-за целочисленности количества этажей в доме);
* *is_apartment* — является ли апартаментами (с object на bool, из-за бинарной природы столбца);
* *balcony* — число балконов (с float64 на int64, из-за целочисленности количества балконов);
* *parks_around3000* — число парков в радиусе 3 км (с float64 на int64, из-за целого количества парков в определённой зоне);
* *ponds_around3000* — число водоёмов в радиусе 3 км (с float64 на int64, из-за целого количества водоёмов в определённой зоне);
* *days_exposition* — число дней от размещения до снятия объявления (с float64 на int64, из-за того, что количество дней является целым числом).

Разберём оставшиеся столбцы.

* *total_images* — число фотографий квартиры в объявлении (сейчас int64, нужно int64);
* *last_price* — цена на момент снятия с публикации (сейчас float64, нужно float64);
* *total_area* — площадь квартиры в квадратных метрах (сейчас float64, нужно float64);
* *first_day_exposition* — дата публикации объявления (сейчас object, нужно datetime64);
* *rooms* — число комнат (сейчас int64, нужно int64);
* *floor* — этаж (сейчас int64, нужно int64);
* *studio* — является ли квартира студией (сейчас bool, нужно bool);
* *open_plan* — есть ли свободная планировка (сейчас bool, нужно bool).

Итого, к первому списку добавляем лишь столбец *first_day_exposition*.

```python
# apts['floors_total'] = apts['floors_total'].astype(int)
# при вызове будет ошибка
```

Из-за наличия пропусков в столбце *floors_total* мы не можем изменить тип данных в нём на целочисленный. Скорее всего, всё-таки, стоит заменить пропуски на неадекватное значение, например, -9999. Тогда конвертация пройдёт успешно. А в дальнешем анализе будем фильтровать эти записи, если будем рассматривать значения этого столбца. Такая ситуация повторится и в столбце *days_exposition*, поэтому в нём сделаем то же самое.

```python
apts['floors_total'] = apts['floors_total'].fillna(-9999)
apts['days_exposition'] = apts['days_exposition'].fillna(-9999)
```

Теперь смело можно менять типы данных.

```python
apts['floors_total'] = apts['floors_total'].astype(int)
apts['is_apartment'] = apts['is_apartment'].astype(bool)
apts['balcony'] = apts['balcony'].astype(int)
apts['parks_around3000'] = apts['parks_around3000'].astype(int)
apts['ponds_around3000'] = apts['ponds_around3000'].astype(int)
apts['days_exposition'] = apts['days_exposition'].astype(int)
apts['first_day_exposition'] = pd.to_datetime(apts['first_day_exposition'], 
                                              format='%Y-%m-%dT%H:%M:%S')
```

Проверим успешность наших замен.

```python
print(apts.info())
```

## Шаг 3. Посчитайте и добавьте в таблицу

### Цена квадратного метра

По очевидной формуле, деля полную стоимость квартиры на её полную площадь, получим цену её квадратного метра.

```python
apts['sqmeter_price'] = apts['last_price'] / apts['total_area']
```

### День недели, месяц и год публикации объявления

Вычленим из даты объявления (столбец *first_day_exposition*) день недели, месяц и год публикации.

```python
apts['weekday_exp'] = apts['first_day_exposition'].dt.weekday # день недели
apts['month_exp'] = apts['first_day_exposition'].dt.month # месяц
apts['year_exp'] = apts['first_day_exposition'].dt.year # год
```

### Этаж квартиры (первый, последний или другой)

Мы должны будем использовать столбец *floors_total* для опознавания последнего этажа. Так как мы заменили пропуски в нём на отрицательное значение, неясно, в какую категорию относить эти объявления. Но таких проблемных записей у нас всего 86, поэтому все их отнесём к категории этажа "другой".

```python
def floor_level(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floors_total < 0:
        return 'другой'
    else:
        if floor == 1:
            return 'первый'
        if floor == floors_total:
            return 'последний'
        return 'другой'

apts['floor_level'] = apts.apply(floor_level, axis=1)
```

Получилось, теперь есть чёткое разделение этажности на три группы.

### Соотношение жилой и общей площади, отношение площади кухни к общей

Чтобы отыскать эти соотношения, нужно лишь поделить значения в столбце *living_area* (и *kitchen_area*, соответственно) на значение в столбце *total_area*. Однако, у меня есть пропуски в столбцах, которые мы делим. В этом нет проблем, просто на выходе снова будет пропуск вместо значения. Главное то, что нет пропусков в знаменателе нашей дроби, столбце, на который мы делим.

```python
apts['living_area_frac'] = apts['living_area'] / apts['total_area']
apts['kitchen_area_frac'] = apts['kitchen_area'] / apts['total_area']
```

Итак, выведем информацию о данных и первые строки обновлённой таблицы.

```python
print(apts.info())
display(apts.head(10))
```

Кажется, всё сделано верно.

### Изучите предложения квартир: для каждой квартиры есть информация о расстоянии до центра. Выделите квартиры в Санкт-Петербурге. Выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Постройте график: он должен показывать, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется — это и будет центральная зона

```python
spb_apts = apts[apts['locality_name'] == 'Санкт-Петербург']
spb_apts = spb_apts.dropna(subset=['cityCenters_nearest']) 
# изучаем только те, где есть информация о расстоянии
spb_apts['km_to_center'] = round(spb_apts['cityCenters_nearest'] / 1000).astype(int)
(spb_apts.pivot_table(index='km_to_center', values='last_price', aggfunc='mean')
        .plot(legend=False)
)
plt.title('')
plt.suptitle('')
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.xlabel('Расстояние до центра (км)')
plt.ylabel('Цена квартиры (млн)')
```

Мы можем заметить, что поведение графика сильно меняется после прохождения отметки в 3 км. Будем считать зону центральной, если расстояние до центра города составляет не более 3 км.

Выделим сегмент квартир в центре.

```python
central_apts = spb_apts[spb_apts['km_to_center'] <= 3]
```

Теперь проанализируем этот набор. Изучим распределение площади квартир в центре и сравним его с распределением по Санкт-Петербургу в общем.

Теперь в обеих группах отыщем факторы, влияющие на стоимость квартиры. Принцип тот же, что и раньше — построить "ящик с усами" для дискретных переменных и диаграмму разброса для непрерывных, а также для всех предложенных столбцов подсчитать корреляцию со столбцом цены квартиры. Сначала узнаем, влияет ли где-то (в центре и/или по всему городу) количество комнат на цену квартиры.

```python
central_apts.boxplot(column='last_price', by='rooms')
plt.ylim(0, 100e6)
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('В центре')
plt.suptitle('')
plt.xlabel('Количество комнат в квартире')
plt.ylabel('Цена квартиры')
plt.show()

spb_apts.boxplot(column='last_price', by='rooms')
plt.ylim(0, 100e6)
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('По всему городу')
plt.suptitle('')
plt.xlabel('Количество комнат в квартире')
plt.ylabel('Цена квартиры')
plt.show()

print(central_apts['rooms'].corr(apts['last_price']))
print(spb_apts['rooms'].corr(apts['last_price']))
```

Мы можем сделать следующий вывод: количество комнат действительно немного влияет на цену квартиры как в центре, так и по всему городу (чем больше комнат, тем выше цена), причём в целом по Санкт-Петербургу зависимость сильнее.

Изучим зависимость цены от того, на каком этаже находится квартира.

```python
central_apts.boxplot(column='last_price', by='floor_level')
plt.ylim(0, 20e6)
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('В центре')
plt.suptitle('')
plt.xlabel('Расположение этажа в доме')
plt.ylabel('Цена квартиры')
plt.show()

spb_apts.boxplot(column='last_price', by='floor_level')
plt.ylim(0, 20e6)
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('По всему городу')
plt.suptitle('')
plt.xlabel('Расположение этажа в доме')
plt.ylabel('Цена квартиры')
plt.show()
```

Итак, из-за типа значений в столбце этажа мы не можем подсчитать корреляцию Пирсона. Однако, по "ящику" мы по-прежнему можем сделать некоторые выводы. В частности, что очевидно, цены на квартиры в центре серьёзно отличаются от среднего по городу. Что важнее, так это то, что ни в центре, ни по городу в целом не очень котируют первый этаж (последний тоже, но в меньшей степени). Но если в среднем по городу квартира на первом этаже ценится меньше квартиры на "другом" этаже примерно на 15 процентов, то в центре эта "скидка" составляет существенные 30 процентов. Пожалуй, это логично, ведь квартиры на первом этаже в центре города означают сосуществование с торговыми точками (и это в лучшем случае). Это мало кому нравится.

Ещё один заявленный фактор, который нуждается в анализе — это расстояние до центра.

```python
central_apts.plot(x='cityCenters_nearest', y='last_price', kind='scatter', alpha=0.3, 
                 ylim=(0, 40e6))
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('В центре')
plt.suptitle('')
plt.xlabel('Расстояние до центра (м)')
plt.ylabel('Цена квартиры')
plt.show()

spb_apts.plot(x='cityCenters_nearest', y='last_price', kind='scatter', alpha=0.3, 
                 ylim=(0, 40e6))
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.title('По всему городу')
plt.suptitle('')
plt.xlabel('Расстояние до центра (м)')
plt.ylabel('Цена квартиры')
plt.show()

print(central_apts['cityCenters_nearest'].corr(apts['last_price']))
print(spb_apts['cityCenters_nearest'].corr(apts['last_price']))
```

Как и ожидается, в самом центре города уже не слишком важно расстояние до географического центра города, ведь 2 километра до него или 3 — это неважно, ведь транспортные службы отлично справляются на таком расстоянии. Коэффициент корреляции же для всего города сразу имеет большее абсолютное значение, то есть зависимость цены от расстояния здесь играет намного большую роль.
