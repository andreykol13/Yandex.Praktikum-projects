## –®–∞–≥ 3. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏ –¥–æ–±–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü—É

### –¶–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞

–ü–æ –æ—á–µ–≤–∏–¥–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ, –¥–µ–ª—è –ø–æ–ª–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã –Ω–∞ –µ—ë –ø–æ–ª–Ω—É—é –ø–ª–æ—â–∞–¥—å, –ø–æ–ª—É—á–∏–º —Ü–µ–Ω—É –µ—ë –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞.

–í—ã—á–ª–µ–Ω–∏–º –∏–∑ –¥–∞—Ç—ã –æ–±—ä—è–≤–ª–µ–Ω–∏—è (—Å—Ç–æ–ª–±–µ—Ü *first_day_exposition*) –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü –∏ –≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏.

### –≠—Ç–∞–∂ –∫–≤–∞—Ä—Ç–∏—Ä—ã (–ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π –∏–ª–∏ –¥—Ä—É–≥–æ–π)

–ú—ã –¥–æ–ª–∂–Ω—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å—Ç–æ–ª–±–µ—Ü *floors_total* –¥–ª—è –æ–ø–æ–∑–Ω–∞–≤–∞–Ω–∏—è –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç—Ç–∞–∂–∞. –¢–∞–∫ –∫–∞–∫ –º—ã –∑–∞–º–µ–Ω–∏–ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –Ω—ë–º –Ω–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –Ω–µ—è—Å–Ω–æ, –≤ –∫–∞–∫—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é –æ—Ç–Ω–æ—Å–∏—Ç—å —ç—Ç–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è. –ù–æ —Ç–∞–∫–∏—Ö –ø—Ä–æ–±–ª–µ–º–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π —É –Ω–∞—Å –≤—Å–µ–≥–æ 86, –ø–æ—ç—Ç–æ–º—É –≤—Å–µ –∏—Ö –æ—Ç–Ω–µ—Å—ë–º –∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —ç—Ç–∞–∂–∞ "–¥—Ä—É–≥–æ–π".

```python
def floor_level(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floors_total < 0:
        return '–¥—Ä—É–≥–æ–π'
    else:
        if floor == 1:
            return '–ø–µ—Ä–≤—ã–π'
        if floor == floors_total:
            return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        return '–¥—Ä—É–≥–æ–π'

apts['floor_level'] = apts.apply(floor_level, axis=1)
```

–ü–æ–ª—É—á–∏–ª–æ—Å—å, —Ç–µ–ø–µ—Ä—å –µ—Å—Ç—å —á—ë—Ç–∫–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ —ç—Ç–∞–∂–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç—Ä–∏ –≥—Ä—É–ø–ø—ã.

### –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∂–∏–ª–æ–π –∏ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏, –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π

–ß—Ç–æ–±—ã –æ—Ç—ã—Å–∫–∞—Ç—å —ç—Ç–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è, –Ω—É–∂–Ω–æ –ª–∏—à—å –ø–æ–¥–µ–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ *living_area* (–∏ *kitchen_area*, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ) –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ *total_area*. –û–¥–Ω–∞–∫–æ, —É –º–µ–Ω—è –µ—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –¥–µ–ª–∏–º. –í —ç—Ç–æ–º –Ω–µ—Ç –ø—Ä–æ–±–ª–µ–º, –ø—Ä–æ—Å—Ç–æ –Ω–∞ –≤—ã—Ö–æ–¥–µ —Å–Ω–æ–≤–∞ –±—É–¥–µ—Ç –ø—Ä–æ–ø—É—Å–∫ –≤–º–µ—Å—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è. –ì–ª–∞–≤–Ω–æ–µ —Ç–æ, —á—Ç–æ –Ω–µ—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ –Ω–∞—à–µ–π –¥—Ä–æ–±–∏, —Å—Ç–æ–ª–±—Ü–µ, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–π –º—ã –¥–µ–ª–∏–º.

```python
apts['living_area_frac'] = apts['living_area'] / apts['total_area']
apts['kitchen_area_frac'] = apts['kitchen_area'] / apts['total_area']
```

–ü–æ—á–µ–º—É —ç—Ç–æ –≤–∞–∂–Ω–æ? 
    
- —ç–∫–æ–Ω–æ–º–∏–º —Å–≤–æ—ë –≤—Ä–µ–º—è
- –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ–º –∫–æ–¥
- –¥–µ–ª–∞–µ–º —Ä–∞–±–æ—Ç—É –ø—Ä–µ–∑–µ–Ω—Ç–∞–±–µ–ª—å–Ω–µ–µ
    
–í—ã–±–µ—Ä–∏ –ª—é–±–æ–π —É—á–∞—Å—Ç–æ–∫ –∫–æ–¥–∞, –≥–¥–µ —ç—Ç–æ —Å–¥–µ–ª–∞—Ç—å —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–µ–µ –≤—Å–µ–≥–æ.

<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π.</b> –ü–æ—Å—Ç–∞—Ä–∞–ª—Å—è —Å–¥–µ–ª–∞—Ç—å –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—é, –ø—Ä–∞–≤–¥–∞ –¥–ª—è –º–µ–Ω—è –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω–µ–π –æ–∫–∞–∑–∞–ª—Å—è –º–µ—Ç–æ–¥ —Å–æ–∑–¥–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏. –í–µ–¥—å –º–µ–∂–¥—É –≥—Ä–∞—Ñ–∏–∫–∞–º–∏ —è —Ö–æ—á—É –ø–∏—Å–∞—Ç—å —Ç–µ–∫—Å—Ç (–∞ –≤ —Ü–∏–∫–ª–µ —É–∫–∞–∑—ã–≤–∞—Ç—å –µ–≥–æ –∫–∞–∫ —á–∞—Å—Ç—å –æ—á–µ—Ä–µ–¥–Ω–æ–≥–æ —à–∞–≥–∞ —è –Ω–µ —Ö–æ—Ç–µ–ª, –≥—Ä–æ–º–æ–∑–¥–∫–æ –±—ã–ª–æ –±—ã). –ê —Ç–∞–∫–∂–µ —É –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ —Ö–æ—Ç–µ–ª–æ—Å—å –ø–æ—Ç—Ä–µ–±–æ–≤–∞—Ç—å –±–æ–ª–µ–µ –¥–µ—Ç–∞–ª–∏–∑–∞—Ü–∏–∏, –±–æ–ª—å—à–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ (–Ω–µ 2, –∞ 3, –∫ –ø—Ä–∏–º–µ—Ä—É). –í—Å–µ –ø—Ä–æ–ø–∏—Å—ã–≤–∞—Ç—å –≤ for –Ω–µ –æ—á–µ–Ω—å-—Ç–æ –∏ —Ö–æ—Ç–µ–ª–æ—Å—å. –ù–∞–¥–µ—é—Å—å, —á—Ç–æ —Ç–∞–∫–∞—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è –ø—Ä–∏–µ–º–ª–µ–º–∞.

### –£–±–µ—Ä–∏—Ç–µ —Ä–µ–¥–∫–∏–µ –∏ –≤—ã–±–∏–≤–∞—é—â–∏–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏—è. –û–ø–∏—à–∏—Ç–µ, –∫–∞–∫–∏–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –æ–±–Ω–∞—Ä—É–∂–∏–ª–∏

–ë—ã–ª–æ —Ä–µ—à–µ–Ω–æ, —á—Ç–æ –º—ã –æ—Å—Ç–∞–≤–∏–º –ª–∏—à—å –æ–±—ä—è–≤–ª–µ–Ω–∏—è —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–Ω–µ–π –ø—Ä–æ–¥–∞–∂–∏ –æ—Ç 3 –¥–æ 1000 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
new_apts = apts.query('3 <= days_exposition <= 1000')
print(new_apts['days_exposition'].describe())
```

–ú—ã —É–±—Ä–∞–ª–∏ –æ–∫–æ–ª–æ 3.5 —Ç—ã—Å—è—á –∑–∞–ø–∏—Å–µ–π, —á—Ç–æ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –æ–∫–æ–ª–æ 15 –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –æ—Ç –≤—Å–µ–≥–æ —á–∏—Å–ª–∞. –í–æ–∑–º–æ–∂–Ω–æ, —ç—Ç–æ –ø–æ–∫–∞–∂–µ—Ç—Å—è –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º, –Ω–æ –∑–∞–º–µ—Ç–∏–º, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –æ—Ç—Å–µ—è–Ω–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π –≤–æ–æ–±—â–µ –Ω–µ –∏–º–µ–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —ç—Ç–æ–º –ø–æ–ª–µ, –ª–∏—à—å –ø—Ä–æ–ø—É—Å–∫–∏. –°—Ç—Ä–æ–∫ —Å –¥–∞–Ω–Ω—ã–º–∏ –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –º—ã —É–±—Ä–∞–ª–∏ –ª–∏—à—å –æ–∫–æ–ª–æ 300 —à—Ç—É–∫.

–ü–æ—Å–ª–µ —Ç–∞–∫–æ–≥–æ —Å—Ä–µ–∑–∞ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω—É –≤ —Å—Ç–æ–ª–±—Ü–µ.

```python
print('–ü–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –∫—Ä–∞–π–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ:')
print('–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–¥–∞–∂–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã',
      new_apts['days_exposition'].mean().round(2), '–¥–Ω–µ–π')
print('–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è –ø—Ä–æ–¥–∞–∂–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã',
      new_apts['days_exposition'].median(), '–¥–Ω–µ–π')
```

–°—Ä–µ–¥–Ω–µ–µ —É–º–µ–Ω—å—à–∏–ª–æ—Å—å –Ω–∞ 17 (–∏ —Ç–µ–ø–µ—Ä—å —Ä–∞–≤–Ω–æ 167 –¥–Ω—è–º), –∞ –º–µ–¥–∏–∞–Ω–∞ –ª–∏—à—å –Ω–∞ 3 (—Ç–µ–ø–µ—Ä—å 93 –¥–Ω—è), —á—Ç–æ —Å–Ω–æ–≤–∞ –≥–æ–≤–æ—Ä–∏—Ç –æ –µ—ë —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏.

### –ö–∞–∫–∏–µ —Ñ–∞–∫—Ç–æ—Ä—ã –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –≤–ª–∏—è—é—Ç –Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã? –ò–∑—É—á–∏—Ç–µ, –∑–∞–≤–∏—Å–∏—Ç –ª–∏ —Ü–µ–Ω–∞ –æ—Ç –ø–ª–æ—â–∞–¥–∏, —á–∏—Å–ª–∞ –∫–æ–º–Ω–∞—Ç, —É–¥–∞–ª—ë–Ω–Ω–æ—Å—Ç–∏ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞. –ò–∑—É—á–∏—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç —Ç–æ–≥–æ, –Ω–∞ –∫–∞–∫–æ–º —ç—Ç–∞–∂–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞: –ø–µ—Ä–≤–æ–º, –ø–æ—Å–ª–µ–¥–Ω–µ–º –∏–ª–∏ –¥—Ä—É–≥–æ–º. –¢–∞–∫–∂–µ –∏–∑—É—á–∏—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Ç –¥–∞—Ç—ã —Ä–∞–∑–º–µ—â–µ–Ω–∏—è: –¥–Ω—è –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü–∞ –∏ –≥–æ–¥–∞

–ò–∑—É—á–∏–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –ø–ª–æ—â–∞–¥–∏. –î–ª—è —ç—Ç–æ–≥–æ –ø–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞–∑–±—Ä–æ—Å–∞, –≥–¥–µ –ø–æ –æ—Å–∏ x –æ—Ç–ª–æ–∂–∏–º –æ–±—â—É—é –ø–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã, –∞ –ø–æ –æ—Å–∏ y ‚Äî —Ü–µ–Ω—É –Ω–∞ –Ω–µ—ë.

–ú—ã –ø–æ–º–Ω–∏–º, —á—Ç–æ –æ—Å–Ω–æ–≤–Ω–æ–µ –æ–±–ª–∞–∫–æ –¥–∞–Ω–Ω—ã—Ö –ª–µ–∂–∏—Ç —Ç–∞–º, –≥–¥–µ —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–µ –±–æ–ª—å—à–µ 20 –º–ª–Ω, –∞ –ø–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã –Ω–µ –±–æ–ª–µ–µ 400 –∫–≤. –º–µ—Ç—Ä–æ–≤. –û—Å—Ç–∞–≤–∏–º –ª–∏—à—å —ç—Ç—É —á–∞—Å—Ç—å –≥—Ä–∞—Ñ–∏–∫–∞.

–í–∏–¥–∏–º –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —á—ë—Ç–∫—É—é –ø—Ä—è–º—É—é –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É —Å—Ç–æ–ª–±—Ü–∞–º–∏, —á—Ç–æ —Å–æ–≥–ª–∞—Å—É–µ—Ç—Å—è —Å —Ä–µ–∞–ª—å–Ω–æ—Å—Ç—å—é. –í –∫–∞—á–µ—Å—Ç–≤–µ –µ—â—ë –æ–¥–Ω–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –ø—Ä–∏–≤–µ–¥—ë–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ –º–µ–∂–¥—É –¥–≤—É–º—è –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏.

### –í—ã–±–µ—Ä–∏—Ç–µ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ —ç—Ç–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö. –í—ã–¥–µ–ª–∏—Ç–µ —Å—Ä–µ–¥–∏ –Ω–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è

–ò—Ç–∞–∫, –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å—Ç–æ–ª–±—Ü–µ *locality_name*.

```python
print(apts['locality_name'].value_counts()[:10])
```

–ò—Ç–∞–∫, –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –æ–±—ä—è–≤–ª–µ–Ω–∏–π –∏–∑ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–∞, –¥–µ—Å—è—Ç–∫—É –∑–∞–º—ã–∫–∞–µ—Ç –í—ã–±–æ—Ä–≥.

–ù–∞–π–¥—ë–º —Ü–µ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ –∫–∞–∂–¥–æ–º –∏–∑ —ç—Ç–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤, –Ω–∞–π–¥—ë–º —Å—Ä–µ–¥–∏ –Ω–∏—Ö –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é.

```python
top_towns = apts['locality_name'].value_counts()[:10].index.to_list() # —Å–ø–∏—Å–æ–∫ –ø—É–Ω–∫—Ç–æ–≤
pop_apts = apts.query('locality_name in @top_towns')
(pop_apts.pivot_table(index='locality_name', values='sqmeter_price', aggfunc='mean')
         .sort_values(by='sqmeter_price', ascending=False))
```

–ü–æ–ª—É—á–∏–ª–∏ —Å–ø–∏—Å–æ–∫ —Ü–µ–Ω –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ —Ä–∞–∑–Ω—ã—Ö –º–µ—Å—Ç–∞—Ö, –ø—Ä–∏—á—ë–º –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –º–µ—Ç—Ä –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ—Ç–∏—Ä—É–µ—Ç—Å—è –Ω–∞ —É—Ä–æ–≤–Ω–µ 115 —Ç—ã—Å—è—á, –∞ –≤ –í—ã–±–æ—Ä–≥–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∏–∂–µ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ ‚Äî 58 —Ç—ã—Å—è—á.

### –ò–∑—É—á–∏—Ç–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –∫–≤–∞—Ä—Ç–∏—Ä: –¥–ª—è –∫–∞–∂–¥–æ–π –∫–≤–∞—Ä—Ç–∏—Ä—ã –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–∏ –¥–æ —Ü–µ–Ω—Ç—Ä–∞. –í—ã–¥–µ–ª–∏—Ç–µ –∫–≤–∞—Ä—Ç–∏—Ä—ã –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ. –í—ã—è—Å–Ω–∏—Ç—å, –∫–∞–∫–∞—è –æ–±–ª–∞—Å—Ç—å –≤—Ö–æ–¥–∏—Ç –≤ —Ü–µ–Ω—Ç—Ä. –°–æ–∑–¥–∞–π—Ç–µ —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≤ –∫–∏–ª–æ–º–µ—Ç—Ä–∞—Ö: –æ–∫—Ä—É–≥–ª–∏—Ç–µ –¥–æ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∏–ª–æ–º–µ—Ç—Ä–∞. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫: –æ–Ω –¥–æ–ª–∂–µ–Ω –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å, –∫–∞–∫ —Ü–µ–Ω–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —É–¥–∞–ª—ë–Ω–Ω–æ—Å—Ç–∏ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –≥—Ä–∞–Ω–∏—Ü—É, –≥–¥–µ –≥—Ä–∞—Ñ–∏–∫ —Å–∏–ª—å–Ω–æ –º–µ–Ω—è–µ—Ç—Å—è ‚Äî —ç—Ç–æ –∏ –±—É–¥–µ—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–∞—è –∑–æ–Ω–∞

```python
spb_apts = apts[apts['locality_name'] == '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥']
spb_apts = spb_apts.dropna(subset=['cityCenters_nearest']) 
# –∏–∑—É—á–∞–µ–º —Ç–æ–ª—å–∫–æ —Ç–µ, –≥–¥–µ –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–∏
spb_apts['km_to_center'] = round(spb_apts['cityCenters_nearest'] / 1000).astype(int)
(spb_apts.pivot_table(index='km_to_center', values='last_price', aggfunc='mean')
        .plot(legend=False)
)
plt.title('')
plt.suptitle('')
plt.ticklabel_format(axis='y', style='sci', scilimits=(6,6))
plt.xlabel('–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ (–∫–º)')
plt.ylabel('–¶–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã (–º–ª–Ω)')
```

–ú—ã –º–æ–∂–µ–º –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ –ø–æ–≤–µ–¥–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Å–∏–ª—å–Ω–æ –º–µ–Ω—è–µ—Ç—Å—è –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –æ—Ç–º–µ—Ç–∫–∏ –≤ 8 –∫–º. –ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å –∑–æ–Ω—É —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π, –µ—Å–ª–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–µ –±–æ–ª–µ–µ 8 –∫–º.

```python
print('–°—Ä–µ–¥–Ω—è—è –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤ –≤ –∫–≤–∞—Ä—Ç–∏—Ä–µ –≤ —Ü–µ–Ω—Ç—Ä–µ',
      central_apts['ceiling_height'].mean().round(2))
print('–°—Ä–µ–¥–Ω—è—è –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤ –≤ –∫–≤–∞—Ä—Ç–∏—Ä–µ –≤–æ –≤—Å—ë–º –≥–æ—Ä–æ–¥–µ',
      spb_apts['ceiling_height'].mean().round(2))

print('–ú–µ–¥–∏–∞–Ω–Ω–∞—è –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤ –≤ –∫–≤–∞—Ä—Ç–∏—Ä–µ –≤ —Ü–µ–Ω—Ç—Ä–µ',
      central_apts['ceiling_height'].median())
print('–ú–µ–¥–∏–∞–Ω–Ω–∞—è –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤ –≤ –∫–≤–∞—Ä—Ç–∏—Ä–µ –≤–æ –≤—Å—ë–º –≥–æ—Ä–æ–¥–µ',
      spb_apts['ceiling_height'].median())
```
